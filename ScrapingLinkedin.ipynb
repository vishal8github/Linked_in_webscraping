```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from parsel import Selector
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import random
from time import sleep
```

```python
service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
driver.get('https://www.linkedin.com/login')

with open('login.txt', 'r') as f:
    credentials = f.readlines()
    username, password = credentials

input_element1 = driver.find_element(By.ID, "username")
input_element1.send_keys(username)
sleep(0.5)

input_element2 = driver.find_element(By.ID, "password")
input_element2.send_keys(password + Keys.ENTER)
sleep(11)
```

```python
driver.get("https://www.google.com/search?q=site%3Alinkedin.com%2Fin%2F+AND+%22Python+Developer%22+AND+%22Mumbai%22&sca_esv=593418533&ei=o3eIZa7fG-ru4-EP9eieaA&ved=0ahUKEwjuvKec2aiDAxVq9zgGHXW0Bw0Q4dUDCBA&uact=5&oq=site%3Alinkedin.com")

href_links = []


wait = WebDriverWait(driver, 50).until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="rso"]/div[1]/div/div/div[1]/div/div/span/a')))
for element in wait:
    link = element.get_attribute('href')
    if link:
        href_links.append(link)
    
    # Break the loop once we have collected five links
    if len(href_links) == 5:
        break

wait = WebDriverWait(driver, 50).until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="rso"]/div[2]/div/div/div[1]/div/div/span/a')))
# Collect up to five href links
for element in wait:
    link = element.get_attribute('href')
    if link:
        href_links.append(link)
    
    # Break the loop once we have collected five links
    if len(href_links) == 5:
        break

wait = WebDriverWait(driver, 50).until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="rso"]/div[3]/div/div/div[1]/div/div/span/a')))
# Collect up to five href links
for element in wait:
    link = element.get_attribute('href')
    if link:
        href_links.append(link)
    
    # Break the loop once we have collected five links
    if len(href_links) == 5:
        break

wait = WebDriverWait(driver, 50).until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="rso"]/div[4]/div/div/div[1]/div/div/span/a')))
# Collect up to five href links
for element in wait:
    link = element.get_attribute('href')
    if link:
        href_links.append(link)
    
    # Break the loop once we have collected five links
    if len(href_links) == 5:
        break

wait = WebDriverWait(driver, 50).until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="rso"]/div[5]/div/div/div[1]/div/div/span/a')))
# Collect up to five href links
for element in wait:
    link = element.get_attribute('href')
    if link:
        href_links.append(link)
    
    # Break the loop once we have collected five links
    if len(href_links) == 5:
        break


# Print the extracted href links
for link in href_links:
    print(link)
```

```python
jobdata = []

for x in href_links:
        #get profile URL
    driver.get(x)
    sleep(random.uniform(2.5, 4.9))

    #assigning the source code for the web page to variable sel
    sel = Selector(text=driver.page_source)

    name = sel.xpath('//*[starts-with(@class, "text-heading-xlarge inline t-24 v-align-middle break-words")]/text()').extract_first()
    if name:
        name = name.strip()


    job_title = sel.xpath('//*[starts-with(@class, "text-body-medium break-words")]/text()').extract_first()
    if job_title:
        job_title = job_title.strip()


    try:
        company = driver.find_element(By.XPATH, '//ul[@class="pv-text-details__right-panel"]').text
    except:
        company = 'None'
    if company:
        company = company.strip()


    location = sel.xpath('//*[starts-with(@class, "text-body-small inline t-black--light break-words")]/text()').extract_first()
    if location:
        location = location.strip()


    #validating if the fields exist on the profile
    # name = validate_field(name)        
    # job_title = validate_field(job_title)        
    # company = validate_field(company)              
    # location = validate_field(location)
    # linkedin_url = validate_field(i)   


    print("\n")
    print(f"Name: {name}")             
    print(f"Job Title: {job_title}")             
    print(f"Company: {company}")             
    print(f"Location: {location}")             
    print(f"Linkedin URL: {x}")             
    print("\n")


    data = {
        'Name': name,
        'Job Title': job_title,
        'Company': company,
        'Location': location,
        'Linkedin URL': x
    }

    jobdata.append(data)

df = pd.DataFrame(jobdata)
df.to_csv('Jobdata_linkedin.csv')
```
